# Using Python's datetime

## 1. combine()

```python
import datetime 

# (hours, minutes)
start_time = datetime.time(7,0)

# (year, month, day)
start_date = datetime.date(2019, 8, 28)

# create a datetime object 
start_datetime = datetime.datetime.combine(start_date, start_time)
```

Above we combined *date* and *time* object. We created a time with an hour of 7 and the minute 0. This represents 7 o'clock.

# Finding the difference between dates
- difference between two dates in days
- difference between two dates in weeks
- difference between two dates in months 
- difference between two dates in years 

```python
import pandas as pd 
import numpy as np  
import datetime 
from dateutil.relativedelta import relativedelta
from datetime import date 

# creating dataframe with random dates 
date1 = pd.Series(pd.date_range('2018-1-1 12:00:00', periods=7, freq='M'))
date2 = pd.Series(pd.date_range('2019-3-11 21:45:00', periods=7, freq='W'))

date_df = pd.DataFrame(dict(Start_date = date1, End_date = date2))
date_df
```

### Finding the difference between two dates in days

```python
# Calculates the difference between two dates 
date_df['diff_days'] = date_df['End_date'] - date_df['Start_date']

# converts the difference in terms of days 
date_df['diff_days']= date_df['diff_days'] / np.timedelta64(1,'D')
 
date_df
```

### Finding the difference between two dates in weeks 

```python
# Calculates the difference between two dates 
date_df['diff_weeks'] = date_df.End_date - date_df.Start_date

# converts the difference in terms of weeks 
date_df['diff_weeks'] = date_df.diff_weeks / np.timedelta64(1, 'W')

date_df
```

### Finding the difference between two dates in months 

```python
# Calculates the difference between two dates 
date_df['diff_months'] = date_df.End_date - date_df.Start_date

# converts the difference in terms of weeks 
date_df['diff_months'] = date_df.diff_months / np.timedelta64(1, 'M')

date_df
```

### Finding the difference between two dates in years

```python
# Calculates the difference between two dates 
date_df['diff_years'] = date_df.End_date - date_df.Start_date

# converts the difference in terms of weeks 
date_df['diff_years'] = date_df.diff_years / np.timedelta64(1, 'Y')

date_df
```
